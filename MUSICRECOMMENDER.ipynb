```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
song_df.head()
```

```python
features = ['title', 'release', 'artist_name', 'song_id']
for feature in features:
    song_df[feature] = song_df[feature].fillna('')
```

```python
def combined_features(row):
    return row['title']+" "+row['release']+" "+row['artist_name']+" "+row['song_id']
song_df["combined_features"] = song_df.apply(combined_features, axis =1)
```

```python
song_df.head()
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(song_df["combined_features"])
print("Count Matrix:", count_matrix.toarray())
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
print(cosine_sim)
```

```python
song_user_likes = "U Smile"
def get_index_from_title(title):
    return song_df[song_df.title == title]["index"]
song_index = get_index_from_title(song_user_likes)
```

```python
similar_songs = list(enumerate(cosine_sim[song_index]))
```

```python
sorted_similar_songs = sorted(similar_songs, key=lambda x:x[1], reverse=True)
```

```python
def get_title_from_index(index):
    return song_df[song_df.index == index]["title"].values[0]
i=0
for song in sorted_similar_songs:
    print(get_title_from_index(song[1]))
    i=i+1
    if i>15:
        break
```
